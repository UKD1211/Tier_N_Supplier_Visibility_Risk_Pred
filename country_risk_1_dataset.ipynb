```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_excel('/content/P_NP Country Risk perso assessment.xlsx')
```

```python
df.sample(10)
```

```python
df.info()
```

```python
df = df[['Time','Country Name','Control of Corruption: Estimate [CC.EST]', 'Government Effectiveness: Estimate [GE.EST]', 'Political Stability and Absence of Violence/Terrorism: Estimate [PV.EST]','Rule of Law: Estimate [RL.EST]','Inflation, consumer prices (annual %) [FP.CPI.TOTL.ZG]', 'Unemployment, total (% of total labor force) (modeled ILO estimate) [SL.UEM.TOTL.ZS]', 'Risk premium on lending (lending rate minus treasury bill rate, %) [FR.INR.RISK]']]
```

```python
df
```

```python
df.isnull().sum()
```

```python
df = df.dropna()
```

```python
df.iloc[4,:]
```

```python
df['Country Name'].value_counts()
```

```python
df['Risk premium on lending (lending rate minus treasury bill rate, %) [FR.INR.RISK]'] = df['Risk premium on lending (lending rate minus treasury bill rate, %) [FR.INR.RISK]'].replace('..', np.nan).astype(float)
df.groupby('Country Name')['Risk premium on lending (lending rate minus treasury bill rate, %) [FR.INR.RISK]'].mean()
```

```python
df.drop(columns = ['Risk premium on lending (lending rate minus treasury bill rate, %) [FR.INR.RISK]'], inplace = True)
```

```python
df
```

```python
df = df[(df['Time'] == 2021) | (df['Time'] == 2022) | (df['Time'] == 2023)]
```

```python
df.replace("..", np.nan, inplace=True)
```

```python
df
```

```python
df.isnull().sum()
```

```python
for cols in df.columns[2:]:
  df.loc[:, cols] = df[cols].fillna(df[cols].median())
```

```python
df
```

```python
x1 = df.groupby('Country Name')['Control of Corruption: Estimate [CC.EST]'].mean().values
x2 = df.groupby('Country Name')['Government Effectiveness: Estimate [GE.EST]'].mean().values
x3 = df.groupby('Country Name')['Political Stability and Absence of Violence/Terrorism: Estimate [PV.EST]'].mean().values
x4 = df.groupby('Country Name')['Rule of Law: Estimate [RL.EST]'].mean().values
x5 = df.groupby('Country Name')['Inflation, consumer prices (annual %) [FP.CPI.TOTL.ZG]'].mean().values
x6 = df.groupby('Country Name')['Unemployment, total (% of total labor force) (modeled ILO estimate) [SL.UEM.TOTL.ZS]'].mean().values
```

```python
df_countries = pd.DataFrame(np.column_stack([x1,x2,x3,x4,x5,x6]),
                            columns = df.columns[2:],
                            index = df.groupby('Country Name')['Control of Corruption: Estimate [CC.EST]'].mean().index)
```
